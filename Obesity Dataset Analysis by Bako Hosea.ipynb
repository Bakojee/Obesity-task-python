### Simple Linear Regression

```python
X = Obesity['Age']
y = Obesity['Height']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```

Coefficient for Age:
With the coefficient for Age at 0.0937. It indicates that for each additional year of age, the height is expected to increase by approximately 0.0937 cm. This suggests a positive relationship between age and height; as individuals get older, they tend to be slightly taller.

Statistical Significance:
The p-value for Age is 0.000, which is significantly less than the common threshold of 0.05. This implies that the relationship between age and height is statistically significant. Therefore, age is a meaningful predictor of height in this model.

R-squared Value:
The R-squared value is 0.013. This means that only 1.3% of the variance in height is explained by age. It is quite low, suggesting that while age has a statistically significant effect on height, it explains only a small portion of the variability in height. Other factors likely to play a much larger role.

F-statistic:
The F-statistic is 21.74, with a corresponding p-value of 3.38e-06. This indicates that the model as a whole is statistically significant, meaning that at least one predictor variable (in this case, age) significantly predicts the dependent variable (height).

### PART 6. Multiple Linear Regression

```python
X_multiple = Obesity[['Age', 'Height', 'Frequency_of_Consuming_Vegetables', 
                  'Number_of_Main_Meals_Daily', 'Liquid_Intake_Daily', 
                  'Physical_Excercise', 'Schedule_Dedicated_to_Technology']]
y_multiple = Obesity['Overweight_Obese_Family']

X_multiple = sm.add_constant(X_multiple)
model_multiple = sm.OLS(y_multiple, X_multiple).fit()

print(model_multiple.summary())
```

# PART 6. INTERPRETATION

R-squared Value:
The R-squared value is 0.108, indicating that approximately 10.8% of the variance in Overweight_Obese_Family is explained by the predictors in the model. Therefore, while the model has some explanatory power, a significant portion of the variance remains unexplained.
